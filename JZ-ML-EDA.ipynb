```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Ames_HousePrice.csv'
ames_df = pd.read_csv(file_path)


# Mapping for full neighborhood names
neighborhood_mapping = {
    'Blmngtn': 'Bloomington Heights',
    'Blueste': 'Bluestem',
    'BrDale': 'Briardale',
    'BrkSide': 'Brookside',
    'ClearCr': 'Clear Creek',
    'CollgCr': 'College Creek',
    'Crawfor': 'Crawford',
    'Edwards': 'Edwards',
    'Gilbert': 'Gilbert',
    'IDOTRR': 'Iowa DOT and Rail Road',
    'MeadowV': 'Meadow Village',
    'Mitchel': 'Mitchell',
    'Names': 'North Ames',
    'NoRidge': 'Northridge',
    'NPkVill': 'Northpark Villa',
    'NridgHt': 'Northridge Heights',
    'NWAmes': 'Northwest Ames',
    'OldTown': 'Old Town',
    'SWISU': 'South & West of Iowa State Uni',
    'Sawyer': 'Sawyer',
    'SawyerW': 'Sawyer West',
    'Somerst': 'Somerset',
    'StoneBr': 'Stone Brook',
    'Timber': 'Timberland',
    'Veenker': 'Veenker'
}

# Mapping for full house style names
housestyle_mapping = {
    '1Story': '1 story',
    '1.5Fin': '1.5 story: 2nd level finished',
    '1.5Unf': '1.5 story: 2nd level unfinished',
    '2Story': '2 story',
    '2.5Fin': '2.5 story: 2nd level finished',
    '2.5Unf': '2.5 story: 2nd level unfinished',
    'SFoyer': 'Split Foyer',
    'SLvl': 'Split Level'
}

# Mapping for full kitchen quality names
kitchenqual_mapping = {
    'Ex': 'Excellent',
    'Gd': 'Good',
    'TA': 'Typical/Average',
    'Fa': 'Fair',
    'Po': 'Poor'
}

# Apply the mappings to the DataFrame
ames_df['Neighborhood'] = ames_df['Neighborhood'].map(neighborhood_mapping)
ames_df['HouseStyle'] = ames_df['HouseStyle'].map(housestyle_mapping)
ames_df['KitchenQual'] = ames_df['KitchenQual'].map(kitchenqual_mapping)


# Set plot style
sns.set(style="whitegrid")


# Neighborhood distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=ames_df, x='Neighborhood', order=ames_df['Neighborhood'].value_counts().index, hue='Neighborhood', palette='viridis', legend=False)
plt.title('Distribution of Neighborhoods')
plt.xticks(rotation=90)
plt.show()

# HouseStyle distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=ames_df, x='HouseStyle', order=ames_df['HouseStyle'].value_counts().index, hue='HouseStyle', palette='viridis', legend=False)
plt.title('Distribution of House Styles')
plt.xticks(rotation=45)
plt.show()

# YearBuilt distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=ames_df, x='YearBuilt', kde=True, color='blue')
plt.title('Distribution of Year Built')
plt.xlabel('Year Built')
plt.ylabel('Frequency')
plt.show()

# CentralAir distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=ames_df, x='CentralAir', hue='CentralAir', palette='viridis', legend=False)
plt.title('Distribution of Central Air')
plt.xlabel('Central Air')
plt.ylabel('Frequency')
plt.show()

# KitchenQual distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=ames_df, x='KitchenQual', order=ames_df['KitchenQual'].value_counts().index,hue='KitchenQual', palette='viridis', legend=False)
plt.title('Distribution of Kitchen Quality')
plt.xlabel('Kitchen Quality')
plt.ylabel('Frequency')
plt.show()

# GrLivArea distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=ames_df, x='GrLivArea', kde=True, color='purple')
plt.title('Distribution of Ground Living Area')
plt.xlabel('Ground Living Area (sq ft)')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of the SalePrice
plt.figure(figsize=(10, 6))
sns.histplot(ames_df['SalePrice'], kde=True, color='blue')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

# Relationships with SalePrice
plt.figure(figsize=(12, 6))
sns.boxplot(data=ames_df, x='Neighborhood', y='SalePrice', hue='Neighborhood', palette='viridis', legend=False)
plt.title('Sale Price by Neighborhood')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(12, 6))
sns.scatterplot(data=ames_df, x='YearBuilt', y='SalePrice', hue='YearBuilt', palette='viridis', legend=None)
plt.title('Sale Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=ames_df, x='CentralAir', y='SalePrice',hue='CentralAir', palette='viridis', legend=False)
plt.title('Sale Price by Central Air')
plt.xlabel('Central Air')
plt.ylabel('Sale Price')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=ames_df, x='KitchenQual', y='SalePrice', hue='KitchenQual', palette='viridis', legend=False)
plt.title('Sale Price by Kitchen Quality')
plt.xlabel('Kitchen Quality')
plt.ylabel('Sale Price')
plt.show()


# Visualize the relationship between GrLivArea (Above grade living area) and SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x=ames_df['GrLivArea'], y=ames_df['SalePrice'], color='green')
plt.title('Above Grade Living Area vs Sale Price')
plt.xlabel('Above Grade Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.show()

# Additional Visualizations for HouseStyle

# Sale Price Distribution by House Style
plt.figure(figsize=(10, 6))
sns.violinplot(data=ames_df, x='HouseStyle', y='SalePrice',hue='HouseStyle', palette='viridis', legend=False)
plt.title('Sale Price Distribution by House Style')
plt.xlabel('House Style')
plt.ylabel('Sale Price')
plt.xticks(rotation=45)
plt.show()

# Average Sale Price by House Style
average_price_by_house_style = ames_df.groupby('HouseStyle')['SalePrice'].mean().sort_values()

plt.figure(figsize=(10, 6))
average_price_by_house_style.plot(kind='bar', color='salmon')
plt.title('Average Sale Price by House Style')
plt.xlabel('House Style')
plt.ylabel('Average Sale Price')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Visualize missing values in the dataset
plt.figure(figsize=(12, 8))
sns.heatmap(ames_df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values in the Ames Housing Dataset')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

# Calculate the percentage of missing values for each column
missing_data = ames_df.isnull().mean().sort_values(ascending=False)

# Filter columns with the top 20 highest missing data percentages
top_20_missing = missing_data.head(20)

# Visualize the top 20 columns with the highest missing data percentages
plt.figure(figsize=(12, 8))
top_20_missing.plot(kind='bar', color='salmon')
plt.title('Top 20 Columns with the Highest Percentage of Missing Data')
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Data')
plt.xticks(rotation=45)
plt.show()

# Calculate correlation of each numeric feature with SalePrice
numeric_cols = ames_df.select_dtypes(include=['number']).columns
correlation_with_saleprice = ames_df[numeric_cols].corr()['SalePrice'].sort_values(ascending=False)

# Filter the top 20 columns most correlated with SalePrice
top_20_correlated = correlation_with_saleprice.head(21).index  # Including SalePrice itself

# Create a subset of the dataframe with the top 20 correlated columns
top_20_correlated_df = ames_df[top_20_correlated]

# Visualize missing values for the top 20 correlated columns
plt.figure(figsize=(12, 8))
sns.heatmap(top_20_correlated_df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values in the Top 20 Columns Most Correlated with SalePrice')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
