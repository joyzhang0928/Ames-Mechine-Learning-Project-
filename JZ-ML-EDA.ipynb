```python

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


# Load the dataset
housing = pd.read_csv('Ames_HousePrice.csv', index_col=0)

# Remove columns with missing values completely to simplify
housing.dropna(axis=1, inplace=True)

# Define the target variable
y = housing['SalePrice']
housing.drop('SalePrice', axis=1, inplace=True)

# Identify categorical and numerical columns
categorical_cols = housing.select_dtypes(include=['object']).columns.tolist()
numerical_cols = housing.select_dtypes(exclude=['object']).columns.tolist()

# Preprocessing pipeline for numeric features
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Preprocessing pipeline for categorical features
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Apply transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

housing_preprocessed = preprocessor.fit_transform(housing)

# Convert the preprocessed categorical data back to a DataFrame
cat_encoded_features = preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_cols)
housing_cat_preprocessed = pd.DataFrame.sparse.from_spmatrix(
    preprocessor.named_transformers_['cat'].transform(housing[categorical_cols]),
    columns=cat_encoded_features
)

# Convert the preprocessed numerical data back to a DataFrame
housing_num_preprocessed = pd.DataFrame(
    preprocessor.named_transformers_['num'].transform(housing[numerical_cols]),
    columns=numerical_cols
)

# Summary of transformations
num_original = housing[numerical_cols]
cat_original_counts = housing[categorical_cols].nunique().sum()
cat_transformed_counts = housing_cat_preprocessed.shape[1]

num_original, housing_num_preprocessed, cat_original_counts, cat_transformed_counts
```

```python
# Select a representative numerical variable for visualization
num_var = 'GrLivArea'

# Plot histograms for the selected numerical variable before and after preprocessing
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

# Before preprocessing
housing[num_var].hist(ax=axes[0], bins=30, color='blue', alpha=0.7)
axes[0].set_title('Before Preprocessing (Original)')
axes[0].set_xlabel(num_var)
axes[0].set_ylabel('Frequency')

# After preprocessing
housing_num_preprocessed[num_var].hist(ax=axes[1], bins=30, color='green', alpha=0.7)
axes[1].set_title('After Preprocessing (Standardized)')
axes[1].set_xlabel(num_var)
axes[1].set_ylabel('Frequency')

plt.suptitle(f'Histograms of {num_var} Before and After Preprocessing')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
